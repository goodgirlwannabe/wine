```python
import pandas as pd
df = pd.read_csv('anggur.csv')
df
```

# Total Sulfur Dioxide

```python
# 7 Total Sulfur Dioxide
# Mean
meanTotalSulfurDioxide = df['total sulfur dioxide'].mean()
print ("Mean Total Sulfur Dioxide: ", meanTotalSulfurDioxide)

# Median
medianTotalSulfurDioxide = df['total sulfur dioxide'].median()
print ("Median Total Sulfur Dioxide: ", medianTotalSulfurDioxide)

# Mode
modusTotalSulfurDioxide = df['total sulfur dioxide'].mode()
print ("Modus Total Sulfur Dioxide: ", modusTotalSulfurDioxide)

# Standar Deviasi
standarDeviasiTotalSulfurDioxide = df['total sulfur dioxide'].std()
print ("Standar Deviasi Total Sulfur Dioxide: ", standarDeviasiTotalSulfurDioxide)

# Variance
varianceTotalSulfurDioxide = df['total sulfur dioxide'].var()
print ("Variance Total Sulfur Dioxide: ", varianceTotalSulfurDioxide)

# Range
rangeTotalSulfurDioxide = df['total sulfur dioxide'].max() - df['total sulfur dioxide'].min()
print ("Range Total Sulfur Dioxide: ", rangeTotalSulfurDioxide)

# Minimum
minimumTotalSulfurDioxide = df['total sulfur dioxide'].min()
print ("Minimum Total Sulfur Dioxide: ", minimumTotalSulfurDioxide)

# Maximum
maximumTotalSulfurDioxide = df['total sulfur dioxide'].max()
print ("Maximum Total Sulfur Dioxide: ", maximumTotalSulfurDioxide)

# Quartile
quartileTotalSulfurDioxide = df['total sulfur dioxide'].quantile([0.25,0.5,0.75])
print ("Quartile Total Sulfur Dioxide: ")
print ("     1. Kuartil 25%: ", quartileTotalSulfurDioxide[0.25])
print ("     2. Kuartil 50%: ", quartileTotalSulfurDioxide[0.5])
print ("     3. Kuartil 75%: ", quartileTotalSulfurDioxide[0.75])

# IQR
IQRTotalSulfurDioxide = quartileTotalSulfurDioxide[0.75] - quartileTotalSulfurDioxide[0.25]
print ("IQR Total Sulfur Dioxide: ", IQRTotalSulfurDioxide)

# Skewness
skewnessTotalSulfurDioxide = df['total sulfur dioxide'].skew()
print ("Skewness Total Sulfur Dioxide: ", skewnessTotalSulfurDioxide)

# Kurtosis
kurtosisTotalSulfurDioxide = df['total sulfur dioxide'].kurt()
print ("Kurtosis Total Sulfur Dioxide: ", kurtosisTotalSulfurDioxide)
```
