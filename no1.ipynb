# Fixed Acidity

```python
# mean
meanFixedAcid = df['fixed acidity'].mean()
print("Rata-rata nilai fixed acidity adalah:             ", meanFixedAcid)

# median
medianFixedAcid = df['fixed acidity'].median()
print("Median nilai fixed acidity adalah:                ", medianFixedAcid)

#modus
modusFixedAcid = df['fixed acidity'].mode()
print("Modus nilai fixed acidity adalah:                 ", modusFixedAcid)

# standard deviation
stdFixedAcid = df['fixed acidity'].std()
print("Standar deviasi nilai fixed acidity adalah:       ", stdFixedAcid)

# variance
varFixedAcid = df['fixed acidity'].var()
print("Variansi nilai fixed acidity adalah:              ", varFixedAcid)

# range
rangeFixedAcid = df['fixed acidity'].max() - df['fixed acidity'].min()
print("Range nilai fixed acidity adalah:                 ", rangeFixedAcid)

# minimum
minFixedAcid = df['fixed acidity'].min()
print("Nilai minimum fixed acidity adalah:               ", minFixedAcid)

# maximum
maxFixedAcid = df['fixed acidity'].max()
print("Nilai maximum fixed acidity adalah:               ", maxFixedAcid)

# kuartil
quartileFixedAcid25 = df['fixed acidity'].quantile(0.25)
quartileFixedAcid50 = df['fixed acidity'].quantile(0.50)
quartileFixedAcid75 = df['fixed acidity'].quantile(0.75)
print("Kuartil nilai fixed acidity: ")
print("     1. Kuartil 25% adalah", quartileFixedAcid25)
print("     2. Kuartil 50% adalah", quartileFixedAcid50)
print("     3. Kuartil 75% adalah", quartileFixedAcid75)

# interquartile range
iqrFixedAcid = quartileFixedAcid75 - quartileFixedAcid25
print("Interquartile range nilai fixed acidity adalah:   ", iqrFixedAcid)

# skewness
skewFixedAcid = df['fixed acidity'].skew()
print("Skewness nilai fixed acidity adalah:              ", skewFixedAcid)

# kurtosis
kurtFixedAcid = df['fixed acidity'].kurt()
print("Kurtosis nilai fixed acidity adalah:              ", kurtFixedAcid)
```

Interquartile range adalah selisih dari nilai Quartile 3 (75%) dan Quartile 1 (25%) untuk mengidentifikasi nilai pencilan.

Skewness merupakan kemiringan distribusi data. Skewness dapat bernilai dapat bernilai positif, negatif, serta nol. Atribut yang memiliki nilai skewness positif artinya memiliki lebih banyak data di sebelah kiri daripada di sebelah kanan. Hal ini disebabkan oleh ekor kurva yang berada pada di sebelah kanan, begitu pula sebaliknya.

Kurtosis atau keruncingan merupakan tingkat puncak dari sebuah distribusi yang biasanya diambil secara relatif terhadap suatu distribusi normal. Semakin besar maka semakin runcing kurva tersebut. Nilai nol menunjukkan data normal, nilai positif menunjukkan data semakin homogen, sedangkan  nilai negatif menunjukkan data semakin tidak homogen.

Kondisi ideal merupakan kondisi ketika kurva terdistribusi secara normal, dan dapat terjadi ketika nilai skewness bernilai nol dan nilai kurtosis bernilai 3. Pada atribut di atas, tidak ada kurva yang memiliki distribusi normal, tetapi terdapat atribut yang memiliki kurva normal (nilai skewness nol).
