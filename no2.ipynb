Membuat Visualisasi plot distribusi, dalam bentuk histogram dan boxplot untuk setiap
kolom numerik. Berikan uraian penjelasan kondisi setiap kolom berdasarkan kedua plot
tersebut.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

df = pd.read_csv('anggur.csv')
df
```

# Fixed Acidity

```python
df.hist(column = 'fixed acidity', figsize = (12,6), rwidth = 0.9, bins = 10, color = '#722F37')
# give title
plt.title('Histogram of Fixed Acidity')
# give xlabel
plt.xlabel('Fixed Acidity')
# give ylabel
plt.ylabel('Frequency')
# show plot
plt.show()
```

```python
df.boxplot(column='fixed acidity', figsize = (12,6), meanline = True, showmeans = True)
# give title
plt.title('Boxplot of Fixed Acidity')
# give ylabel
plt.ylabel('Frequency')
# show plot
plt.show()
```

Histogram tersebut menunjukkan bahwa distribusi fixed acidity terlihat condong ke arah kanan (positively skewed). Selain itu, boxplot menunjukkan bahwa terdapat beberapa data outlier yang berada di rentang 3-4 dan 10-12 dan median terlihat di sekitar 7.1, dengan kuartil pertama sekitar 6.3 dan kuartil ketiga tepat di 8, serta nilai minimum di sekitar 3.3 dan nilai maksimum berada pada sekitar 11.5
