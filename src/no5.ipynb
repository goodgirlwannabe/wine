Melakukan test hipotesis 2 sampel,

```python
import scipy.stats as st
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind

df = pd.read_csv('anggur.csv')
df.head()
```

### <b> a. Data kolom fixed acidity dibagi 2 sama rata: bagian awal dan bagian akhir kolom. Benarkah rata-rata kedua bagian tersebut sama? </b>

$H_{0} $: nilai rata-rata bagian awal kolom sama dengan nilai rata-rata bagian akhir kolom $(\mu_{awal} - \mu_{akhir} = 0) \\ $
$H_{1} $: nilai rata-rata bagian awal kolom tidak sama dengan nilai rata-rata bagian akhir kolom $(\mu_{awal} - \mu_{akhir} \neq 0) \\ $
Dengan $\alpha = 0.05$

Melakukan pengujian two tailed test untuk mengecek bagian kiri dan kanan dengan $z < -z_{\alpha / 2}$ dan $z > z_{\alpha / 2}$ dengan rumus $$z = \frac{(\bar{x_{1}} - \bar{x_{2}}) - (\mu_{1} - \mu_{2})}{\sqrt{\sigma^2_{1}/n_{1} + \sigma^2_{2}/n_{2}} }$$

$H_{0}$ ditolak ketika $ z > z_{\alpha / 2} $ atau $ z < -z_{\alpha / 2} $ dan $p < \alpha $

```python
delta = 0
alpha = 0.05

len_data = len(df)
len_data_per_2 = len_data/2
df_awal = df['fixed acidity'][:int(len_data_per_2)]
df_akhir = df['fixed acidity'][int(len_data_per_2):]
z, p = ztest(df_awal, df_akhir, value=delta, alternative='two-sided')

z_alpha = st.norm.ppf(1-alpha/2)

print (f'Nilai Z adalah: {round(z,4)}')
print (f'Nilai Z_alpha adalah: {round(z_alpha,4)}')
print (f'Nilai P adalah: {round(p,4)}')
plt.subplot(1,2,1)
df_awal.plot(kind='box', title='Data Awal')
plt.subplot(1,2,2)
df_akhir.plot(kind='box', title='Data Akhir')
plt.show()
```

Dapat disimpulkan bahwa <b> rata-rata bagian awal dan akhir kolom fixed acidity adalah tidak sama </b> dikarenakan nilai z yang kurang dari $z_{\alpha / 2}$

### <b> b. Data kolom chlorides dibagi 2 sama rata: bagian awal dan bagian akhir kolom. Benarkah rata-rata bagian awal lebih besar daripada bagian akhir sebesar 0.001? </b>

$H_{0} $: nilai rata-rata bagian awal kolom sama dengan nilai rata-rata bagian akhir kolom $(\mu_{awal} - \mu_{akhir} = 0.001) \\ $
$H_{1} $: nilai rata-rata bagian awal kolom tidak sama dengan nilai rata-rata bagian akhir kolom $(\mu_{awal} - \mu_{akhir} \neq 0.001) \\ $
Dengan $\alpha = 0.05$

Melakukan pengujian two tailed test untuk mengecek bagian kiri dan kanan dengan $z < -z_{\alpha / 2}$ dan $z > z_{\alpha / 2}$ dengan rumus $$z = \frac{(\bar{x_{1}} - \bar{x_{2}}) - (\mu_{1} - \mu_{2})}{\sqrt{\sigma^2_{1}/n_{1} + \sigma^2_{2}/n_{2}} }$$

$H_{0}$ ditolak ketika $ z > z_{\alpha / 2} $ atau $ z < -z_{\alpha / 2} $ dan $p < \alpha $

```python
delta = 0.15
alpha = 0.05

len_data = len(df)
len_data_per_2 = len_data/2
df_awal = df['chlorides'][:int(len_data_per_2)]
df_akhir = df['chlorides'][int(len_data_per_2):]
z, p = ztest(df_awal, df_akhir, value=delta, alternative='two-sided')

z_alpha = st.norm.ppf(1-alpha/2)

print (f'Nilai Z adalah: {round(z,4)}')
print (f'Nilai Z_alpha adalah: {round(z_alpha,4)}')
print (f'Nilai P adalah: {round(p,4)}')
plt.subplot(1,2,1)
df_awal.plot(kind='box', title='Data Awal')
plt.subplot(1,2,2)
df_akhir.plot(kind='box', title='Data Akhir')
plt.show()
```

Berdasarkan perhitungan diatas, dapat disimpulkan bahwa <b> rata-rata bagian awal lebih besar dari bagian akhir kolom chlorides sebesar 0.001 </b>

### <b> c. Benarkah rata-rata sampel 25 baris pertama kolom Volatile Acidity sama dengan rata-rata 25 baris pertama kolom Sulphates ? </b>
$H_{0} $: nilai rata-rata bagian awal kolom sama dengan nilai rata-rata bagian akhir kolom $(\mu_{awal} - \mu_{akhir} = 0) \\ $
$H_{1} $: nilai rata-rata bagian awal kolom tidak sama dengan nilai rata-rata bagian akhir kolom $(\mu_{awal} - \mu_{akhir} \neq 0) \\ $
Dengan $\alpha = 0.05$

Melakukan pengujian two tailed test untuk mengecek bagian kiri dan kanan dengan $z < -z_{\alpha / 2}$ dan $z > z_{\alpha / 2}$ dengan rumus $$z = \frac{(\bar{x_{1}} - \bar{x_{2}}) - (\mu_{1} - \mu_{2})}{\sqrt{\sigma^2_{1}/n_{1} + \sigma^2_{2}/n_{2}} }$$

$H_{0}$ ditolak ketika $ z > z_{\alpha / 2} $ atau $ z < -z_{\alpha / 2} $ dan $p < \alpha $

```python
delta = 0
alpha = 0.05

df_va_awal = df['volatile acidity'].head(25)
df_sul_awal = df['sulphates'].head(25)
z, p = ttest_ind(df_va_awal, df_sul_awal, equal_var=False)

z_alpha = st.norm.ppf(1-alpha/2)

print (f'Nilai Z adalah: {round(z,4)}')
print (f'Nilai Z_alpha adalah: {round(z_alpha,4)}')
print (f'Nilai P adalah: {round(p,4)}')
plt.subplot(1,2,1)
df_va_awal.plot(kind='box', title='Data 25 Volatile Acidity Awal')
plt.subplot(1,2,2)
df_sul_awal.plot(kind='box', title='Data 25 Sulphate Awal')
plt.show()
```

Dapat disimpulkan bahwa <b> rata-rata 25 baris awal volatile acidity tidak sama dengan 25 baris awal sulphates </b>

### <b> d. Bagian awal kolom residual sugar memiliki variansi yang sama dengan bagian akhirnya? </b>
$H_{0} $: variansi awal kolom residual sugar memiliki nilai yang sama dengan bagian akhirnya $(\sigma^2_{1} = \sigma^2_{2}) \\ $
$H_{1} $: variansi awal kolom residual sugar memiliki nilai yang berbeda dengan bagian akhirnya $(\sigma^2_{1} \neq \sigma^2_{2}) \\ $
Dengan $\alpha = 0.05$

Melakukan pengujian two tailed test untuk mengecek bagian kiri dan kanan dengan $f > f_{\alpha / 2}(v_{1}, v_{2})$ dengan rumus $$f = \frac{s^2_{1}}{s^2_{2}}$$

$H_{0}$ ditolak ketika $ f > f_{\alpha / 2}(v_{1}, v_{2}) $ atau $ f < f_{1 - \alpha / 2}(v_{1}, v_{2}) $ dan $p < \alpha $

```python
alpha = 0.05
len_data = len(df)
len_data_per_2 = len_data/2
df_awal = df['residual sugar'].head(int(len_data_per_2))
df_akhir = df['residual sugar'].tail(int(len_data_per_2))
v1 = len(df_awal) - 1
v2 = len(df_akhir) - 1

f = df_awal.var()/df_akhir.var()
f_awal = st.f.ppf(1-alpha, v1, v2)
f_akhir = st.f.ppf(alpha, v1, v2)

p = 1 - st.f.cdf(f, v1, v2)

print (f'Nilai F adalah: {round(f,4)}')
print (f'Nilai F_awal adalah: {round(f_awal,4)}')
print (f'Nilai F_akhir adalah: {round(f_akhir,4)}')
print (f'Nilai P adalah: {round(p,4)}')
plt.subplot(1,2,1)
df_awal.plot(kind='box', title='Data Awal')
plt.subplot(1,2,2)
df_akhir.plot(kind='box', title='Data Akhir')
plt.show()
```

Berdasarkan hasil diatas, dapat disimpulkan bahwa <b> bagian awal dan bagian akhir kolom residual sugar memiliki variansi yang berbeda </b>

### <b> e. Proporsi nilai setengah bagian awal alcohol yang lebih dari 7, adalah lebih besar daripada, proporsi nilai yang sama di setengah bagian akhir alcohol? </b>

$H_{0} $: proporsi nilai setengan bagian awal alcohol yang lebih dari 7 sama dengan proposi bagian akhir alcohol $(p_{1} = p_{2}) \\ $
$H_{1} $: proporsi nilai setengan bagian awal alcohol yang lebih dari 7 berbeda dengan proposi bagian akhir alcohol $(p_{1} \neq p_{2}) \\ $
Dengan $\alpha = 0.05$

Uji statistik dilakukan dengan one tailed test untuk mengecek bagian kanan dengan $ z > z_{\alpha} $
Untuk menghitung nilai  z dengan  rumus $$z = \frac{\hat{p_{1}} - \hat{p_{2}}}{\sqrt{\hat{p}\hat{q} (1/n_{1} + 1/n_{2})} }$$
dengan nilai $\hat{p}$ $$ \hat{p} = \frac{x_{1} + x_{2}}{n_{1} + n_{2}}, \hat{q} = 1 - \hat{p}$$

$H_{0}$ ditolak ketika $ z > z_{\alpha} $ dan $p < \alpha $

```python
delta = 0
alpha = 0.05

len_data = len(df)
len_data_per_2 = len_data/2
df_awal = df['alcohol'][:int(len_data_per_2)]
df_akhir = df['alcohol'][int(len_data_per_2):]

z, p = proportions_ztest([len(df_awal) > 7, len(df_akhir)> 7], [len(df_awal), len(df_akhir)], alternative='two-sided')

z_alpha = st.norm.ppf(1-alpha/2)

print (f'Nilai Z adalah: {round(z,4)}')
print (f'Nilai Z_alpha adalah: {round(z_alpha,4)}')
print (f'Nilai P adalah: {round(p,4)}')
plt.subplot(1,2,1)
df_awal.plot(kind='box', title='Data Awal')
plt.subplot(1,2,2)
df_akhir.plot(kind='box', title='Data Akhir')
plt.show()
```

Dapat disimpulkan bahwa <b> proporsi nilai setengan bagian awal alcohol yang lebih dari 7 sama dengan proposi bagian akhir alcohol </b>
