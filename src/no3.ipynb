Menentukan setiap kolom numerik berdistribusi normal atau tidak. Gunakan normality test
yang dikaitkan dengan histogram plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sp
import seaborn as sns

df = pd.read_csv('anggur.csv')
df
```

Normality test dapat dilakukan dengan melakukan pemanggilan fungsi normaltest dari library scipy. Implementasi test ini berdasar pada D'Agostino-Pearson Test yang juga sering disebut sebagai Omnibus D'Agostino. 

D'Agostino-Pearson Test merupakan salah satu metode untuk menguji normalitas sampel data. Data dihitung berdasarkan skewness dan kurtosis data.

Perhitungan D'Agostino-Pearson Test dapat menggunakan rumus berikut.
$$K^2 = Z_{s}^2 + Z_{k}^2 $$


$ Z_{s}^2 $ merupakan Z-score skewness dan $ Z_{k}^2 $ merupakan Z-score kurtosis. Apabila hipotesis nol terbukti, $K^2$ dapat diaproksimasi terdistribusi chi-squared dengan derajat kebebasan 2.

Hipotesis nol $H_{0}$ merupakan data yang berasal dari populasi dengan distribusi normal. $H_{0}$ dapat diuji dengan membandingkan nilai $ \alpha $ yang ditetapkan dengan p-value yang didapat dari normaltest. $H_{0}$ diterima ketika p-value lebih kecil dari $ \alpha $. 

Untuk mempermudah pembuktian hasil normality test, ditampilkan histogram distribusi data dengan memanggil fungsi histplot yang diimport dari library seaborn.

# Fixed Acidity

```python
stat, p = sp.normaltest(df['fixed acidity'])
a = 0.05
print(f"Nilai p ialah: {p}")
if p < a:
        print('Hipotesis nol ditolak, data fixed acidity tidak berdistribusi normal')
else:
        print('Hipotesis nol diterima, data fixed acidity berdistribusi normal')
sns.histplot(df['fixed acidity'])
plt.show()
```
